```python
import pandas as pd
df = pd.read_csv('premier.csv')
```

```python
df.head()
```

```python
company_name = 'Thoughtworks'
company_df = df[df['company'] == company_name]
```

```python
import matplotlib.pyplot as plt

company_df['date'] = pd.to_datetime(company_df['date'])

# Sort the DataFrame by the 'date' column
company_df = company_df.sort_values('date')

# Plot the line chart
plt.plot(company_df['date'], company_df['certifications'])

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Certifications')
plt.title(f'AWS Certifications points of {company_name} over time')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python
company_names = ['Thoughtworks', 'Accenture', 'Deloitte']

# Create a figure and axis objects
fig, ax = plt.subplots()

# Iterate over each company
for company_name in company_names:
    # Filter the DataFrame for the current company
    company_df = df[df['company'] == company_name]

    # Convert the 'date' column to datetime if it's not already in datetime format
    company_df['date'] = pd.to_datetime(company_df['date'])

    # Sort the DataFrame by the 'date' column
    company_df = company_df.sort_values('date')

    # Plot the line chart for the current company
    ax.plot(company_df['date'], company_df['certifications'], label=company_name)

# Set labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Certifications')
ax.set_title('Certifications Comparison of Companies over Time')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Add a legend
ax.legend()

# Display the chart
plt.show()
```
